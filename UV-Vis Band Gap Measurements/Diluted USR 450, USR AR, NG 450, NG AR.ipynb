```python
from kiethley import *
import numpy as np
import matplotlib.pyplot as plt
```

# Read in UV-Vis Data

```python
KBr_1 = spectrum('10-17-24 KBr 1.txt')
KBr_2 = spectrum('10-17-24 KBr 2.txt')
KBr_3 = spectrum('10-17-24 KBr 3.txt')

USR_450_1 = spectrum('10-17-24 USR 450 1.txt')
USR_450_2 = spectrum('10-17-24 USR 450 2.txt')

KBr_4 = spectrum('10-17-24 KBr 1.txt')
USR_AR_1 = spectrum('10-17-24 USR AR 1.txt')

KBr_8 = spectrum('10-17-24 KBr 8.txt')
NG_AR_2 = spectrum('10-17-24 NG AR 2.txt')

KBr_10 = spectrum('10-17-24 KBr 10.txt')
NG_450_1 = spectrum('10-17-24 NG 450 1.txt')
```

# Plot USR 450 Diluted Data

```python
KBr_1.plotspectrum(Label = 'KBr 1') #Had the lights turn on
KBr_2.plotspectrum(Label = 'KBr 2')
KBr_3.plotspectrum(Label = 'KBr 3')

USR_450_1.plotspectrum(Label = 'USR 450 1')
USR_450_2.plotspectrum(Label = 'USR 450 2')
```

```python
USR_450_2.KM_transform(KBr_2)
USR_450_2.tauc_transform()
USR_450_2.plottauc()

plt.title('Tauc Plot USR 450')
USR_450_2.tauc_fit([2.6,3],[3.5,3.9])
```

```python
USR_AR_1.KM_transform(KBr_4)
USR_AR_1.tauc_transform()
USR_AR_1.plottauc()

plt.title('Tauc Plot USR AR')
USR_AR_1.tauc_fit([2.6,3],[3.5,3.9])
```

```python
NG_AR_2.KM_transform(KBr_8)
NG_AR_2.tauc_transform()
NG_AR_2.plottauc()

plt.title('Tauc Plot NG AR')
NG_AR_2.tauc_fit([2.6,3],[3.2,3.9])
```

```python
NG_450_1.KM_transform(KBr_10)
NG_450_1.tauc_transform()
NG_450_1.plottauc()

plt.title('Tauc Plot USR AR')
NG_450_1.tauc_fit([2.6,3],[3.3,3.9])
```

```python
plt.subplot(2, 1, 2)
USR_450_2.tauc_fit([2.6,3],[3.3,3.9],label = 'USR 450, Eg: 3.33eV', color = 'royalblue')
USR_AR_1.tauc_fit([2.7,2.8],[3.3,3.9],label = 'USR AR, Eg: 3.34eV', color = 'green')
NG_AR_2.tauc_fit([2.6,3],[3.2,3.9],label = 'NG AR, Eg: 3.26eV', color = 'red')
NG_450_1.tauc_fit([2.6,3],[3.3,3.9],label = 'NG 450, Eg: 3.26eV', color = 'mediumpurple')


#USR_450_4.plottauc(Label = 'USR 450, Eg: 3.27 eV')
#BC_PB_1.plottauc(Label = 'BC PB, Eg: 3.26')
#USR_AR_1.plottauc(Label = 'USR AR, Eg: 3.25 eV')
#BC_400_1.plottauc(Label = 'BC 400, Eg: 3.22 eV')
#BC_1.plottauc(Label = 'BC, Eg: 3.01 eV')

plt.legend()
plt.xlim(2.5,4.35)
plt.ylim(0,5)

plt.subplot(2, 1, 1)
plt.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
plt.plot(USR_450_2.wavelength,USR_450_2.KM/0.477, label = 'USR 450, Eg: 3.33eV')
plt.plot(USR_AR_1.wavelength,USR_AR_1.KM/0.477, label = 'USR AR, Eg: 3.34eV')
plt.plot(NG_AR_2.wavelength,NG_AR_2.KM/5.56, label = 'NG AR, Eg: 3.26eV')
plt.plot(NG_450_1.wavelength,NG_450_1.KM/5.56, label = 'NG 450, Eg: 3.26eV')
#plt.title('KM Plots of Photocatalysts Catalysts')

plt.legend()
plt.xlabel('Wavelength (nm)')
plt.ylabel('KM Absorbance')
plt.xlim(285,496)
```
