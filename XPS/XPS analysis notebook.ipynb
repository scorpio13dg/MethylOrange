# Define Classes

```python
class spectrum:
    def __init__(self, filepath, source = 'Mg'):
        self.kinetic_energy = []
        self.cps = []
        self.read_file(filepath)
        if source == 'Mg':
            self.binding_energy = 1253.6 - self.kinetic_energy
        elif source == 'Al':
            self.binding_energy = 1486.7 - self.kinetic_energy
        else:
            raise ValueError('x-ray source not recognized')
        self.take_second_derivative()
    
    def read_file(self, filepath):
        with open(filepath, 'r') as f:
            while True:
                newline = f.readline().split()
                if len(newline) == 0:
                    f.close()
                    break
                self.kinetic_energy.append(float(newline[0]))
                self.cps.append(float(newline[1]))
        self.kinetic_energy = np.array(self.kinetic_energy)
        self.cps = np.array(self.cps)
    
    def plot(self):
        plt.plot(self.binding_energy, self.cps)
        plt.xlabel('Binding Energy (eV)')
        plt.ylabel('Intensity (counts per second)')
        plt.xlim(self.binding_energy[0], self.binding_energy[-1])
        
    def take_second_derivative(self):
        print('second derivative nyi')
```
