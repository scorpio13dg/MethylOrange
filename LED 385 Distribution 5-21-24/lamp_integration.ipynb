```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
spectrum = pd.read_csv("halogen_lamp_2_I++.txt", delim_whitespace=True)
```

```python
plt.plot(spectrum.iloc[0:160, 0])
```

```python
I = spectrum.iloc[0:160, 0]
wavelength = spectrum.iloc[0:160, 1]
```

```python
band_gap = 1240/3.2 #band gap in nanometers
print(band_gap)
```

```python
#return index at which point photons fall below the band gap energy
for index, l in enumerate(wavelength):
    if l > band_gap:
        print(index, l)
        break
```

```python
print(f"{np.sum(I[0:29])/np.sum(I):.3%} of photons emitted by the tungsten halogen lamp are 3.2eV or higher")
```

```python
plt.plot(wavelength, I/1E8, label = "spectrum")
plt.xlabel("wavelength (nm)")
plt.ylabel("voltage (V)")
plt.title("tungsten filament halogen lamp")
plt.vlines(band_gap, 0, 1.2, color = "Red", linestyles = "dashed", label = "3.2 eV band gap")
plt.legend()
```
